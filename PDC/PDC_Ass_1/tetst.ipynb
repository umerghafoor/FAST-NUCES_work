```python
import matplotlib.pyplot as plt
import pandas as pd

# Read CSV data with column name cleaning
data = pd.read_csv('01_test.csv')
data.columns = data.columns.str.strip()  # Remove any leading/trailing spaces

# Separate data by affinity
affinity_0 = data[data["Affinity"] == 0]
affinity_1 = data[data["Affinity"] == 1]
```

```python
affinity_1
```

```python
affinity_0
```

```python
# Plotting
plt.figure(figsize=(8, 5))
plt.plot(affinity_0["threads"], affinity_0["processing_time"], marker='o', linestyle='-', label="Processing Time (Affinity = 0)")
plt.plot(affinity_1["threads"], affinity_1["processing_time"], marker='s', linestyle='-', label="Processing Time (Affinity = 1)")

# Labels and title
plt.xlabel("Number of Threads")
plt.ylabel("Processing Time (s)")
plt.title("Processing Time and ECU vs Threads with and without Affinity")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# Plotting
plt.figure(figsize=(8, 5))
plt.plot(affinity_0["threads"], affinity_0["ECU"], marker='^', linestyle='--', label="ECU (Affinity = 0)")
plt.plot(affinity_1["threads"], affinity_1["ECU"], marker='d', linestyle='--', label="ECU (Affinity = 1)")

# Labels and title
plt.xlabel("Number of Threads")
plt.ylabel(" ECU (%)")
plt.title("Processing Time and ECU vs Threads with and without Affinity")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
